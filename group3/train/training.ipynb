```python
import pandas as pd
import os
```

```python
#load mwh data
```

```python
print(os.getcwd())
data = pd.read_csv('../../data/Avtice-losses.csv', header=1)
print(data.head(5))
```

```python
#convert losses from kwh to mwh
data['MWh']=data['kWh']/1000
```

```python
#convert quarter resolution to hourly resolution
data['Zeitstempel'] = pd. to_datetime(data['Zeitstempel']) 
data['time_start'] = data['Zeitstempel']-pd.Timedelta(minutes=15)
```

```python
data['hour'] = data['time_start'].apply(lambda x: x.strftime('%Y-%m-%d %H'))
```

```python
data_hour = data.groupby(['hour'])['MWh'].sum().reset_index()
```

```python
# Remove the first row (index 0)
data_hour = data_hour.drop(0)

# Reset the index if needed
data_hour.reset_index(drop=True, inplace=True)
mwhdata = data_hour
```

```python
display(mwhdata)
```

```python
###load Temperature data
```

```python
# Load the data
data = pd.read_csv("../../data/Forecast-temperature.csv")

# Convert the 'datetime' column to datetime format
data['datetime'] = pd.to_datetime(data['datetime'])

# Check if the data is already in 6-hour intervals
if (data['datetime'].diff().dt.total_seconds() / 3600).median() == 6:
    # If the data is in 6-hour intervals, leave the datetime column as it is
    hourly_data = data
else:
    # If the data is not in 6-hour intervals, resample it to hourly intervals
    hourly_data = data.resample('H', on='datetime').mean().reset_index()

    # Fill missing values in other columns with the values from the initial 6-hourly data
    columns_to_fill = hourly_data.columns.difference(['datetime'])
    hourly_data[columns_to_fill] = hourly_data[columns_to_fill].fillna(method='ffill', limit=6)
    # Replace '6' with the appropriate number if you want to fill more or fewer hours

# Print the resulting hourly data
temperaturedata = hourly_data
#display(temperaturedata)
#temperaturedata.to_csv('your_file.csv', index=False)
```

```python
temperaturedata['datetime'] = pd.to_datetime(temperaturedata['datetime'])
temperaturedata['datetime'] = temperaturedata['datetime'].apply(lambda x: x.strftime('%Y-%m-%d %H'))
temperaturedata.rename(columns={'datetime': 'hour'}, inplace=True)
```

```python
display(temperaturedata)
```

```python
### load renewable data
```

```python
renewabledata = pd.read_csv("../../data/Forecast-renewable-generation.csv")
```

```python
# Remove the first row (index 0)
renewabledata = renewabledata.drop(0)

# Reset the index if needed
renewabledata.reset_index(drop=True, inplace=True)
```

```python
## Data Preparation
```

```python
## Merge DF
```

```python
merged_df = pd.merge(mwhdata, ntcdata, on='hour', how='inner')
merged_df = pd.merge(merged_df, renewabledata, on='hour', how='inner')
merged_df = pd.merge(merged_df, temperaturedata, on='hour', how='inner')
```

```python
display(merged_df)
```

```python
## create lags (the 7 day data as columns)
```

```python
# Create lag features for the past 7 days
for i in range(1, 169):
    merged_df[f'temperature_fore_ch_lag_{i}'] = merged_df['temperature_fore_ch'].shift(i)
    merged_df[f'temperature_fore_fr_lag_{i}'] = merged_df['temperature_fore_fr'].shift(i)
    merged_df[f'temperature_fore_it_lag_{i}'] = merged_df['temperature_fore_it'].shift(i)
    merged_df[f'temperature_fore_de_lag_{i}'] = merged_df['temperature_fore_de'].shift(i)
    
    merged_df[f'wind_fore_it [MW]_lag_{i}'] = merged_df['wind_fore_it [MW]'].shift(i)
    merged_df[f'wind_fore_de [MW]_lag_{i}'] = merged_df['wind_fore_de [MW]'].shift(i)
    merged_df[f'solar_fore_it [MW]_lag_{i}'] = merged_df['solar_fore_it [MW]'].shift(i)
    merged_df[f'solar_fore_de [MW]_lag_{i}'] = merged_df['solar_fore_de [MW]'].shift(i)
    
    merged_df[f'IT_CH_lag_{i}'] = merged_df['IT_CH'].shift(i)
    merged_df[f'FR_CH_lag_{i}'] = merged_df['FR_CH'].shift(i)
    merged_df[f'DE_CH_lag_{i}'] = merged_df['DE_CH'].shift(i)
    merged_df[f'AT_CH_lag_{i}'] = merged_df['AT_CH'].shift(i)
    merged_df[f'CH_IT_lag_{i}'] = merged_df['CH_IT'].shift(i)
    merged_df[f'CH_FR_lag_{i}'] = merged_df['CH_FR'].shift(i)
    merged_df[f'CH_DE_lag_{i}'] = merged_df['CH_DE'].shift(i)
    merged_df[f'CH_AT_lag_{i}'] = merged_df['CH_AT'].shift(i)
    
    merged_df[f'MWh_lag_{i}'] = merged_df['MWh'].shift(i)
# Drop rows with NaN values (first 7 rows)
merged_df.dropna(inplace=True)
```
