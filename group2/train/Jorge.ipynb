```python
active_losses = pd.read_csv('../data/Avtice-losses.csv', skiprows=1)

# Convert the 'Datetime' column to a datetime object
active_losses['Zeitstempel'] = pd.to_datetime(active_losses['Zeitstempel'])

```

```python
active_losses['Zeitstempel'].dt.date
```

```python
# Group the data by both day and hour components of the 'Datetime' column and aggregate the 'Value' column
hourly_daily_aggregated = active_losses.groupby([active_losses['Zeitstempel'].dt.date, active_losses['Zeitstempel'].dt.hour])['kWh'].sum().reset_index()

# Rename the columns for clarity
hourly_daily_aggregated.columns = ['Date', 'Hour', 'kWh']

# Display the aggregated data
print(hourly_daily_aggregated)
```
